Random kép létrehozása

```python
random = np.empty(shape = (434, 434, 3))
```

```python
random = np.empty(shape = (600, 434, 3))

for i in range(600):
  for j in range(434):
    for h in range(3):
      random[i, j, h] = rnd.randint(0, 256)
```

```python
print(random.shape)
print(type(random))
print(type(alpaka))
```

```python
plt.imshow(random)
```

Közös ábrázolás

```python
fig = plt.figure(figsize=(10, 7))

rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(tok)
plt.axis('off')
plt.title("Tök")
  
fig.add_subplot(rows, columns, 2)
plt.imshow(tudta)
plt.axis('off')
plt.title("Tudta?")

fig.add_subplot(rows, columns, 3)
plt.imshow(alpaka)
plt.axis('off')
plt.title("Két szép állat")
  
fig.add_subplot(rows, columns, 4)
plt.imshow(random)
plt.axis('off')
plt.title("Egy kortárs alkotás, Ismeretlen szerző")
```
